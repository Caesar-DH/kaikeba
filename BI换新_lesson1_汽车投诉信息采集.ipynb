### sqrt(10)

```python
def prescribe(num,eps):
    i = 0
    while 1:
       
        if i**2 == num:
            return i
        if i**2 >num:       
            low = i-1
            height = i
            break 
        i+=1
        
    for ind in range(eps):
        mid = (low+height)/2
#         print(mid)
        if mid*mid > num:
            height = mid
        elif  mid*mid == num:
            print(1)
            return mid
        else:
            low = mid
    return mid
```

### 汽车投诉信息采集：

```python
from bs4 import BeautifulSoup
import requests as re
```

```python
def get_page_content(url):
    html = re.get(url,headers={'User-Agent': 'Mozilla/5.0 (iPhone; CPU iPhone OS 11_0 like Mac OS X) AppleWebKit'})
    content = html.text    
    soup = BeautifulSoup(content)   
    return soup
def get_url(base,page):
    return [base.format(x+1) for x in range(page)]
```

```python
urls = get_url('http://www.12365auto.com/zlts/0-0-0-0-0-0_0-0-0-0-0-0-0-{}.shtml',20)
print(urls)
```

```python
# 投诉编号，投诉品牌，投诉车系，投诉车型，问题简述，典型问题，投诉时间，投诉状态
number = []
brand = []
series = []
model = []
briefIntro = []
typicalProblem =[]
time = []
status =[]
for url in urls:
    page = get_page_content(url)
    table = page.table
    for sibling in table.tr.next_siblings:
        number.append(sibling.find_all('td')[0].text)
        brand.append(sibling.find_all('td')[1].text)
        series.append(sibling.find_all('td')[2].text)
        model.append(sibling.find_all('td')[3].text)
        briefIntro.append(sibling.find_all('td')[4].text)
        typicalProblem.append(sibling.find_all('td')[5].text)
        time.append(sibling.find_all('td')[6].text)
        status.append(sibling.find_all('td')[7].text) 
print(number,brand,series,model,briefIntro,typicalProblem,time,status)
```
