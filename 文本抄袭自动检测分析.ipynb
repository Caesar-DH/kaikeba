### Thinking1  什么是监督学习，无监督学习，半监督学习

监督学习任务：从已标记的训练数据来训练模型。 主要分为：分类任务、回归任务、序列标注。

无监督学习任务：从未标记的训练数据来训练模型。主要分为：聚类任务、降维。

半监督学习任务：用大量的未标记训练数据和少量的已标记数据来训练模型。

### Thinking2 K-means中的k值如何选取


手肘法。随k的增大，样本划分会更加精细，每个簇的聚合程度会逐渐提高，那么SSE自然会逐渐变小。而当k到达真实聚类数时，再增加k所得到的聚合程度回报会迅速变小，所以SSE的下降幅度会骤减，然后随着k值的继续增大而趋于平缓。所以找斜率最大的点，就是聚类数量。

### Thinking3 随机森林采用了bagging集成学习，bagging指的是什么

Bagging 是一种随机有放回的算法。

### Thinking4 主动学习和半监督学习的区别是什么

主动学习：机器可以自己选择学习的样本，它可以使用较少的训练取得更好的效果。需要人工介入，模型主动向worker提供数据。

半监督学习：指在训练数据十分稀少的情况下，利用没有标签的数据，提高模型效果的方法。


## 文本抄袭自动检测分析:

```python
import pandas as pd
import numpy as np
import jieba
```

```python
# 数据加载
new = pd.read_csv('sqlResult.csv',encoding='gb18030')
# new.head()
# 删除缺失值
new = new.dropna(subset=['content'])
print(new.shape)
```

```python
# 停用词
with open('chinese_stopwords.txt','r', encoding='utf-8') as file:
    stopwords = [i[:-1] for i in file.readlines()]

# 分词
def split_test(text):
    text = text.replace(' ','').replace('\n','')
    text2 = jieba.cut(text)
    res = ' '.join([w for w in text2 if w not in stopwords])
    return res
```

```python
split_test(new.iloc[0].content)
```

```python
corpus = list(map(split_test,[str(i) for i in new.content]))
print(corpus)
```

```python
# 文本特征
from sklearn.feature_extraction.text import CountVectorizer,TfidfTransformer
contV = CountVectorizer(encoding='gb18030',min_df=0.015)
tfidftrans = TfidfTransformer()
countvector = contV.fit_transform(corpus)
tfidf = tfidftrans.fit_transform(countvector)
print(tfidf.shape)
```

```python
# 标记是否为自己新闻
label = list(map(lambda source: 1 if '新华社' in str(source) else 0 ,new.source ))
print(label)
```

```python
# 数据切分 训练
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score,precision_score,recall_score

X_train,X_text,y_train,y_test = train_test_split(tfidf.toarray(),label,test_size=0.3)

model = MultinomialNB()
model.fit(X_train,y_train)

y_predict = model.predict(X_text)
print("准确率:",accuracy_score(y_test,y_predict))
print("精确率:",precision_score(y_test,y_predict))
print("召回率:",recall_score(y_test,y_predict))
```

```python
# 使用模型进行预测
prediction = model.predict(tfidf.toarray())
labels = np.array(label)
compare_news_index = pd.DataFrame({'prediction':prediction,'labels':labels})
copy_news_idnex = compare_news_index[(compare_news_index['prediction'] == 1)&(compare_news_index['labels']==0)].index

# 实际为新华社的新闻
xinhuashe_news_index = compare_news_index[(compare_news_index['labels']==1)].index

print("可能是抄袭的条数：",len(copy_news_idnex))
```

```python
from sklearn.preprocessing import Normalizer
from sklearn.cluster import KMeans

noremalizer = Normalizer()
scaled_array = noremalizer.fit_transform(tfidf.toarray())

# 聚类
kmeans = KMeans(n_clusters = 25)
k_labels = kmeans.fit_predict(scaled_array)
print(k_labels)
```

```python
# 创建class
from collections import defaultdict
id_class = {index:class_ for index,class_ in enumerate(k_labels)}
class_id = defaultdict(set)
for index,class_ in id_class.items():
    if index in xinhuashe_news_index.tolist():
        class_id[class_].add(index)
```

```python
# 查找相关文章
from sklearn.metrics.pairwise import cosine_similarity

def find_text(cpindex,top=10):
    dest_dict = {i:cosine_similarity(tfidf[cpindex],tfidf[i]) for i in class_id[id_class[cpindex]]}
    return sorted(dest_dict.items(),key = lambda x:x[1][0],reverse=True)[:top]
```

```python
cpindex = 3352
print(cpindex in xinhuashe_news_index)
print("是否在抄袭列表里：",cpindex in copy_news_idnex)
```

```python
slmilar_list = find_text(cpindex)
print(slmilar_list)
print('怀疑抄袭内容：\n',new.iloc[cpindex].content)
similar2 = slmilar_list[0][0]
print("相似原文：\n",new.iloc[similar2].content)
```
